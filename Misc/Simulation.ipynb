```python
import matplotlib.pyplot as plt
import numpy as np


#Num of simulations
N_sim = 1000
#Probability of winning 
P_win = 0.55
#Amount bet each move
M_bet = 0.10
#Starting amount of money
M_start = 100
#Precision to count as zero
Eps = 0.1

#Actual simulation
M = np.zeros(N_sim)
M[0] = M_start
Rand = np.random.random_sample((N_sim,))


for i in range(1,N_sim):
    if Rand[i] < P_win:
        #Win situation
        M[i] = M[i-1]*(1+M_bet)
    else:
        #Loose situation
        M[i] = M[i-1]*(1-M_bet)
    if M[i]<Eps:
        #Setting account balance to zero (not allowing to bet small values)
        M[i]=0

#Plotting
plt.plot(M)
plt.ylabel('Account balance')
plt.show()
```
